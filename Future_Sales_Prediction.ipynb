```python
# Reading data into dataframe
df=pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
df=df.iloc[0:, 1:53]
```

```python
# Removing column 51 from the data
# y has the train data, x has the
y = df['W51']
x = df[[col for col in df if col != 'W51']]
a = x.values.tolist()
```

```python
# Applying LSTM encoding data with window size 3
l1=[]
output=[]
train=[]
l4=[]
for i in range(0,len(a),1):
    for j in range(0,len(a[0])-3,1):
        l1=[]
        l1.append(a[i][j])
        l1.append(a[i][j+1])
        l1.append(a[i][j + 2])
        output.append(a[i][j+3])
        train.append(l1)

l5=[]
test=[]
for i in range(0,len(a),1):

    for j in range(50,51,1):

        l5 = []
        l5.append(a[i][j-2])
        l5.append(a[i][j-1])
        l5.append(a[i][j])
        test.append(l5)
```

```python
# Applying ridge regression to the data created using LSTM encoding
ridge_mod = Ridge(alpha=0.5, normalize=True).fit(train, output)
```

```python
#Predicting W51 values
ypred = ridge_mod.predict(test)
```

```python
#calculating MSE for the predicted data and the actual data
mse = mean_squared_error(y, ypred)


for i in ypred:
    print(i)
print(len(ypred))
print('MSE:',mse)
```
